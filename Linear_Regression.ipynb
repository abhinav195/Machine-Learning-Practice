```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
diabetes = datasets.load_boston()
# print(diabetes)
#dict_keys(['data', 'target', 'DESCR', 'feature_names', 'data_filename', 'target_filename'])
# print(diabetes.keys())
```

```python
diabetes_x = diabetes.data[:, np.newaxis, 2]
#print(diabetes_x)
```

```python
diabetes_x_train = diabetes_x[:-130]
diabetes_x_test = diabetes_x[-130:]
```

```python
diabetes_y_train = diabetes.target[:-130]
diabetes_y_test = diabetes.target[-130:]
# print(diabetes_y_train)
```

```python
model = linear_model.LinearRegression()
model.fit(diabetes_x_train, diabetes_y_train)
diabetes_y_predicted = model.predict(diabetes_x_test)
# diabetes_y_predicted
```

```python
print("Mean Squared Error is :", mean_squared_error(diabetes_y_test, diabetes_y_predicted))
```

```python
print("Mean Absolute Error is :",mean_absolute_error(diabetes_y_test, diabetes_y_predicted))
```

```python
print("Coefficient : ", model.coef_)
print("Intercept : ", model.intercept_)
```

```python
plt.scatter(diabetes_x_train, diabetes_y_train)
plt.show()
plt.scatter(diabetes_x_test, diabetes_y_test)
plt.show()
plt.scatter(diabetes_x_test, diabetes_y_predicted)
plt.show()
```
