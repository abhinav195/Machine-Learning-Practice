```python
pred_var=['texture_mean', 'perimeter_mean','smoothness_mean','compactness_mean','symmetry_mean']
```

```python
train,test= train_test_split(data, test_size=0.3)
print(train.shape)
print(test.shape)
```

```python
x_train= train[pred_var]
x_test= test[pred_var]
y_train= train.diagnosis
y_test=test.diagnosis
```

```python
logmodel= LogisticRegression()
logmodel.fit(x_train,y_train)
y_pred=logmodel.predict(x_test)
```

```python
acc=metrics.accuracy_score(y_pred,y_test)
print(acc)
```

```python
clf=DecisionTreeClassifier(random_state=0)
cross_val_score(clf, x_train,y_train, cv=10)
clf.fit(x_train, y_train,sample_weight=None, check_input=True, X_idx_sorted=None)
clf.get_params(deep=True)
clf.predict(x_test,check_input=True)
clf.predict_log_proba(x_test)
clf.predict(x_test,check_input=True)
print(clf.score(x_test, y_test, sample_weight=None))
```
